```python
def dummy_cat(df2, col):
    '''
    INPUT:
    df - the dataframe where col is stored
    col - the categorical column you want to dummy (as a string)
    OUTPUT:
    df - the dataframe with the added columns
         for dummy variables using 1, 0, -1 coding
    '''
    for idx, val_0 in enumerate(df[col].unique()):
        if idx + 1 < df2[col].nunique():            
            df2[val_0] = df2[col].apply(lambda x: 1 if x == val_0 else 0)
        else:    
            df2[val_0] = df2[col].apply(lambda x: -1 if x == val_0 else 0)
            for idx, val_1 in enumerate(df[col].unique()):
                if idx + 1 < df2[col].nunique():
                    df2[val_1] = df2[val_0] + df2[val_1]
                else:
                    del df2[val_1]
    return df2

new_df = dummy_cat(df2, '결과표현방식') # Use on style
new_df.head(10)
```

```python
new_df['intercept'] = 1

lm = sm.OLS(new_df['조회수'], new_df[['intercept', 'B', 'C']])
results = lm.fit()
results.summary()
```
